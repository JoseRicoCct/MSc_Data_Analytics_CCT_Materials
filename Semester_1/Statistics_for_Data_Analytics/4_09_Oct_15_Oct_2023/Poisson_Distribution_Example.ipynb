```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_parameter = 3  # Average rate of events

# Generate random numbers from a Poisson distribution
data = poisson.rvs(mu=lambda_parameter, size=1000)

# Plot a histogram of the generated data
plt.hist(data, bins=20, density=True, alpha=0.6, color='b', label='Poisson PMF')

# Calculate the Poisson probability mass function
x = np.arange(0, 20)
pmf = poisson.pmf(x, mu=lambda_parameter)
plt.vlines(x, 0, pmf, colors='r', lw=2, alpha=0.6, label='Poisson PMF')

plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.legend()
plt.title('Poisson Distribution')
plt.show()
```
