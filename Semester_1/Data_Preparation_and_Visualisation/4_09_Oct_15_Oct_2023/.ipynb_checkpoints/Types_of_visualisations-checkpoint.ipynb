```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('penguins')
```

```python
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm');
```

```python
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm',hue='species');
```

```python
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm',hue='species',size='sex');
```

Adding alpha value to see through.

Add ; at the end of the code to get rid off the messages.

```python
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm',hue='species',size='sex', alpha=0.7);
```

```python
plt.figure(figsize=(9,5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm',hue='species',size='sex', alpha=0.7);
```

```python
plt.figure(figsize=(9,5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm',hue='species',size='sex', alpha=0.7)

plt.legend(loc='upper right',bbox_to_anchor=(1.2,1));
```

```python
sns.set_palette("rainbow")
plt.figure(figsize=(9,5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm',hue='species',size='sex', alpha=0.7)

plt.legend(loc='upper right',bbox_to_anchor=(1.2,1));
```
